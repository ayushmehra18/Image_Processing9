```python
# To import various libraries

import math
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to read 'palmleaf1.pgm' file which is in P2 format.

def readpgm(name):
    with open(name) as f:
        lines = f.readlines()

    # Ignores commented lines
    for l in list(lines):
        if l[0] == '#':
            lines.remove(l)

    # Makes sure it is ASCII format (P2)
    assert lines[0].strip() == 'P2' 

    # Converts data to a list of integers
    data = []
    for line in lines[1:]:
        data.extend([int(c) for c in line.split()])

    return (np.array(data[3:]),(data[1],data[0]),data[2])

data = readpgm('palmleaf1.pgm')
image1_array = np.reshape(data[0],data[1])

plt.imshow(image1_array, cmap = 'gray')
plt.title('PALMLEAF1')
plt.axis('off')
plt.show()
```

```python
# Array data related to image palmleaf1.pgm

A,B = image1_array.shape
print(f"Shape of the image array is : {A} , {B}\n")
N = image1_array.size
print("Total number of elements in image array : N =",N,"\n")
L = image1_array.max()
print("Maximum intensity in image array : L =", L)
```

```python
# Calculating frequency of all the intensities present in the image palmleaf1.pgm

F = []
for k in range (0,(L+1)):
    c = 0
    for i in range (0,A):
        for j in range (0,B):
            if (image1_array[i,j] == k):
                c = c + 1    
    F.append(c)
    
F = np.array(F)
```

```python
# To plot a graph between frequency of intensity v/s intensity.

p = F.size
x=np.arange(0,p,1)
plt.plot(x, F)
plt.xlabel('Intensity (i)') 
plt.ylabel('F(i)') 
plt.title('F(i) v/s i')
plt.grid()
plt.show() 
```

```python
# Calculation of between class variance.

variance_b = []

for t in range (0,(L+1)):
    N1 = 0
    for i in range (0,A):
        for j in range (0,B):
            if (image1_array[i,j] <= t):
                N1 = N1 + 1
    
    N2 = N - N1

    # mean of first class
    mu1 = 0
    for i in range (0,(t+1)):
        mu1 = mu1 + (i*(F[i]/N1))
        
    # mean of second class
    mu2 = 0
    for i in range ((t+1),(L+1)):
        mu2 = mu2 + (i*(F[i]/N2))
    
    # Total mean of the entire image
    mu_t = 0
    for i in range (0,(L+1)):
        mu_t = mu_t + (i*(F[i]/N))

    sigma_b_sq = ((mu1-mu_t)*(mu1-mu_t)*(N1/N)) + ((mu2-mu_t)*(mu2-mu_t)*(N2/N))
    
    variance_b.append(sigma_b_sq)
    
variance_b = np.array(variance_b)
```

```python
# To find the Optimal Threshold.

t_star = variance_b.argmax()
print("Threshold = ",t_star)
```

```python
# Converting into binary image

binry = np.zeros((A,B))
for i in range (0,A):
    for j in range (0,B):
        if (image1_array[i,j] > t_star):
            binry[i,j] = 1
```

```python
# To display binary image

plt.imshow(binry, cmap='gray')
plt.title('Binary of PALMLEAF1')
plt.axis('off')
plt.show()
```
